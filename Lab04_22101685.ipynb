```python
#Assignment Part 1
class Patient:#node
  #write a constructor
  def __init__ (self, id=None, name=None, age=None, bloodgroup=None, next=None, prev=None):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = next
    self.prev = prev


```

```python
class WRM:
  #write a constructor
  def __init__(self):
    self.head = Patient()
    self.head.next = self.head
    self.head.prev = self.head
    self.tail = self.head


  def registerPatient(self,id, name, age, bloodgroup):
    #To Do
    new_patient = Patient(id, name, age, bloodgroup, self.head, self.tail)
    print(f"The Patient {name} has been registered Successfully. His ID is {id}")
    self.tail.next = new_patient
    self.head.prev = new_patient
    self.tail = self.tail.next



  def servePatient(self):
    #To Do
    if self.head.next != self.head:
      name = self.head.next.name
      self.head.next = self.head.next.next
      self.head.next.next.prev = self.head
      self.tail = self.head.prev
      print(f"{name} has been served.")
    else:
      print("None to Serve")


  def showAllPatient(self):
    #To Do
    if self.head.next != self.head:
      biboron = self.head.next
      while biboron != self.head:
        print(f"Name : {biboron.name} ID : {biboron.id}")
        biboron = biboron.next
    else:
      print("No patient to show!")


  def canDoctorGoHome(self):
    #To Do
    if self.head.next == self.head:
      return "Yes"
    else:
      return "No. You've got a duty to attend!"




  def cancelAll(self):
    #To Do

    if self.head.next != self.head:
      self.head.next=self.head
      self.head.prev=self.head
      print("The appointment has been successfully cancelled")
    else:
      print("no patient so you can go on")



  def ReverseTheLine(self):
    #To Do
    if self.head.next == self.head:
      print("No patient to reverse!")
    else:
      temp = self.head.next
      self.head.next = self.head.prev
      self.head.prev = self.head.next
      while temp != self.head:
        temp.next = temp.prev
        temp.prev = temp.next
        temp = temp.next
      print("Successfully reversed")



```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    e.next = None
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

```python
def diamond_count(stack,string):
  #TO DO
  count = 0
  for i in string:

    if i == "<" :
      stack.push(i)

    elif i == ">" :

      if stack.isEmpty() == True:
        pass

      else:
        stack.pop()
        count += 1
  return count


print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```
